```python
import pandas as pd
import numpy as np
df = pd.read_csv('D:/Leads.csv')
```

```python
Y = df['Converted']
num_col = ['TotalVisits','Page Views Per Visit','Total Time Spent on Website']
cat_col = ['Lead Origin','Lead Source','Last Activity','Last Notable Activity','Specialization','City','What is your current occupation','What matters most to you in choosing a course']
df1 = df[num_col]
df2 = df[cat_col]
X = pd.concat([df1,df2],axis=1)
```

```python
from sklearn.impute import SimpleImputer
from sklearn.metrics import accuracy_score,roc_auc_score
imputer = SimpleImputer(missing_values=np.nan, strategy='mean')
imputer = imputer.fit(X[num_col])
  
X[num_col] = imputer.transform(X[num_col])
```

```python
imputer = SimpleImputer(missing_values=np.nan, strategy='most_frequent')
imputer = imputer.fit(X[cat_col])
  
X[cat_col] = imputer.transform(X[cat_col])
```

```python
from sklearn.model_selection import train_test_split,GridSearchCV
from sklearn.preprocessing import StandardScaler
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size=0.3)
```

```python
scaler = StandardScaler()
X_train[num_col] = scaler.fit_transform(X_train[num_col])
X_test[num_col] = scaler.fit_transform(X_test[num_col])
```

```python
X_train_cat = X_train[cat_col]
X_test_cat = X_test[cat_col]
X_train_cat['train'] = 1
X_train_cat['test'] = 0
X_test_cat['train'] = 0
X_test_cat['test'] = 1
```

```python
X_cat_comb = pd.concat([X_train_cat,X_test_cat])
X_cat_comb = pd.get_dummies(X_cat_comb,columns=cat_col,drop_first=True)
X_cat_comb
```

```python
X_train_cate = X_cat_comb[X_cat_comb['train'] ==1]
X_test_cate = X_cat_comb[X_cat_comb['test'] ==1]
```

```python
X_train_cate.drop('train',axis=1,inplace=True)
X_train_cate.drop('test',axis=1,inplace=True)
```

```python
X_test_cate.drop('train',axis=1,inplace=True)
X_test_cate.drop('test',axis=1,inplace=True)
```

```python
X_train_fin = pd.concat([X_train[num_col],X_train_cate],axis=1)
X_train_fin
```

```python
X_test_fin = pd.concat([X_test[num_col],X_test_cate],axis=1)
X_test_fin
```

```python
from sklearn.ensemble import RandomForestClassifier
rf = RandomForestClassifier(n_estimators=50,max_depth=8)
li = list(range(1,100))
param = {'n_estimators':li}
grid = GridSearchCV(estimator=rf,param_grid=param,cv=5)
grid.fit(X_train_fin,Y_train)
```

```python
grid.best_params_
```

```python
rf = RandomForestClassifier(n_estimators=46,min_samples_split=5)
rf.fit(X_train_fin,Y_train)
```

```python
pred = rf.predict(X_test_fin)
accuracy = accuracy_score(Y_test,pred)
auc_score = roc_auc_score(Y_test,pred)

print(accuracy)
print(auc_score)
```
